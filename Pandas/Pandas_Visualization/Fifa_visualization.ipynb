```python
rc = fifa["Release Clause"].to_list()
actual_values = []
for i in rc:
    if type(i) == str:
        if i[-1] == "M":
            actual_values.append(float(i[1:-1])*1000000)
        elif i[-1] == "K":
            actual_values.append(float(i[1:-1])*1000.0)
    else:
            actual_values.append(0)
```

```python
fifa["RC"] = actual_values
fifa
```

```python
fifa["newRC"] = fifa["Release Clause"].replace([""],[""] , regex= True)
fifa
```

```python
fifa["newcleanRC"] = fifa["Release Clause"].replace(["", "M", "K"],["", "", ""] , regex= True)
fifa
```

```python
def representRC(cell):
    if type(cell) == str:
        if cell[-1] == "M":
            return float(cell[:-1])*1000000.0
        elif cell[-1] == "K":
            return float(cell[:-1])*1000.0
        else:
            return float(cell)
    else:
        return 0.0,
```

```python
fifa.newRC = fifa.newRC.apply(representRC)
fifa
```
