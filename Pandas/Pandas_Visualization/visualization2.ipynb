```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,40]
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend()
plt.show()
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,85,27]



plt.bar(company,revenue, label="Revenue")
addlabels(company, revenue)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,6))
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,85,27]
profit=[40,7,34,12]
xpos = np.arange(len(company)) # numpyarray : 0 , 1 , 2, 3

plt.bar(company,revenue, width=0.4, label="Revenue") # x = 0, x = 0.4
plt.bar(xpos-0.4,profit, width=0.4,label="Profit")# x = -.4  , x = 0
# 0 - 0.4 = -.4 , .6 , 1.6 , 2.6
plt.xticks(xpos,company)
plt.ylabel("Revenue(Bln)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

## pie chart

```python
# pie chart
exp_vals = [1400,600,300,300,250] #expenses
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.0f%%',radius=2.0 )
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%0.4f%%',radius=2.0 )
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, autopct='%1.2f%%' )
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.4f%%',radius=1.75,explode=[0,0,0.5,0,0])
plt.show() # formatting
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0.1,0.3,0.1,0.1,0.1])
plt.legend(bbox_to_anchor =(1.1, 1))
plt.show()
```
