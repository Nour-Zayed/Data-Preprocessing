## Reshape pandas dataframe using melt

```python
import pandas as pd
df = pd.read_csv("/content/weather (1).csv")
df
```

```python
df[["day", "chicago"]]
```

```python
df[["day", "berlin"]]
```

```python
melted = pd.melt(df, id_vars=["day"], var_name='city', value_name='temprature')
melted
```

```python
city = melted [ melted["city"] == "berlin" ]
city
```

```python
city = city.drop("city", axis=1)
city
```

```python
city.rename({"temprature" : "berlin" },axis=1 , inplace=True)
city
```
